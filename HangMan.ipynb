```python
#imports the words from a bank
import requests

word_site = "https://raw.githubusercontent.com/Xethron/Hangman/master/words.txt"

response = requests.get(word_site)
wordL = response.content.splitlines()

#decodes the byte format
def remW(w):
    return w.decode('utf-8')

#maps the function to all words items in the list
wordL=list(map(remW,wordL))
```

```python
#produces random word
import random
def ranW(x):
    return random.choice(x)


#produces a list for enumerated index's
def duplicates(lst, item):
    return [i for i, x in enumerate(lst) if x == item]
```
